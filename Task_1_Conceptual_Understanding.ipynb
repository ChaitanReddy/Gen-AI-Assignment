## Task 1: Conceptual Understanding

**What is the difference between 'Love' and 'love' in NLP?**

 In NLP, 'Love' and 'love' are usually treated as distinct tokens due to capitalization, unless a text normalization step like lowercasing is applied.

**What happens if stopwords are not removed?**

 Not removing stopwords can increase data volume, dilute meaningful terms and potentially negatively impact model performance by overemphasizing common, less informative words.

**Provide two real-world scenarios where removing stopwords can be harmful.**

**Sentiment Analysis**: Removing words like "not" can reverse the intended sentiment (e.g., "not happy").
**Search Engines/Information Retrieval**: Precise phrase searches (e.g., "to kill a mockingbird") can yield less accurate results if stopwords are removed from queries or indexes.

**Explain the difference between stemming and lemmatization.**

Stemming is a heuristic process that chops word endings to find a root, which may not be a real word (e.g., 'running' -> 'runn').

Lemmatization is a more linguistic process that reduces words to their base dictionary form (lemma), which is always a valid word (e.g., 'running' -> 'run').


```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [t for t in tokens if len(t) > 2 or t in ['no','not']]

    return tokens, ' '.join(tokens)
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in test_sentences:
    tokens, clean = preprocess_text(text)
    results.append(tokens)
    print(text)
    print(tokens)
    print(clean)
    print()
```

```python
for tokens in results:
    if tokens:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = sum(len(t) for t in tokens)/total
        print(total, unique, round(avg_len,2))
```

```python
from collections import Counter

all_tokens = [t for sub in results for t in sub]
counter = Counter(all_tokens)

print(counter.most_common(10))
print(sorted(counter.items(), key=lambda x: x[1])[:5])
```

```python
def full_pipeline(text_list):
    tokens_all = []
    clean_sentences = []
    for text in text_list:
        tokens, clean = preprocess_text(text)
        tokens_all.extend(tokens)
        clean_sentences.append(clean)
    return {"tokens": tokens_all, "clean_sentences": clean_sentences}
```
